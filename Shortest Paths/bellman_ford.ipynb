```python
def bellman_ford(G, s):
    d = [999999] * len(G)
    d[s] = 0
    for j in range(len(G)-1):
        for u in G:
            u_index = G.index(u)
            for v in u:
                v_index = v[0]-1
                d[v_index] = min(d[v_index], d[u_index] + v[1])
    return d
```
